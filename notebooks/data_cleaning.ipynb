```python
import pandas as pd
```

```python
# Leemos ficheros con su correspondiente codificacion 
cliclar = pd.read_csv('../data/raw/clicar_raw.csv', encoding='ansi')
Francia = pd.read_csv('../data/raw/autoscout24_france_raw.csv', encoding='ansi')
flexicar = pd.read_csv('../data/raw/flexicar_raw.csv', encoding='utf_8')
ocasion_plus = pd.read_csv('../data/raw/ocasionplus_raw.csv', encoding='ansi')
Holanda= pd.read_csv('../data/raw/autoscout24_netherlands_raw.csv', encoding='ansi')
Alemania = pd.read_csv('../data/raw/mobile_de_raw.csv', encoding='ansi')
autocasion= pd.read_csv('../data/raw/autocasion_raw.csv', encoding='utf8')
```

Limpieza csv Francia

```python
Francia['kilometros']=Francia['kilometros'].map(lambda x: int("".join(str(x).split()[:-1])))  
type(Francia.loc[1,'modelo']) #Vemoss que la columna modelo cada celda es str
Francia['modelo']=Francia['modelo'].map(lambda x: x.split()[0].upper()) #Nos quedamos solo con el modelo 
Francia['precio']=Francia['precio'].map(lambda x: int(x*1000) if "." in str(x)  else int(x)) 
Francia['marca']=Francia['marca'].map(lambda x: x.upper()) #Escrbimos la marca en mayuscula 
Francia['pais']='Francia'

Francia.to_csv('../data/processed/cars_france.csv', index=False) #Pasamos a un nuevo csv el limpio
```

Limpieza clicar

```python
cliclar['kilometros']=cliclar['kilometros'].map(lambda x: float(x[:-2])) #Limpiamos los kilometros
cliclar['kilometros']=cliclar['kilometros'].map(lambda x: int(x*1000) if "." in str(x)  else int(x)) #Pasamos a int los kilometros
cliclar['modelo']=cliclar['modelo'].map(lambda x: x.split()[0].upper()) #Nos quedamos solo con el modelo 
cliclar['precio']=cliclar['precio'].map(lambda x: int(1000*float(x[:-1])))
cliclar['marca']=cliclar['marca'].map(lambda x: x.upper()) #Escrbimos la marca en mayuscula 



```

Modificacion página flexicar


```python
flexicar.rename(columns={'a�o': 'año'}, inplace=True) #Renombramos una column a
flexicar['modelo']=flexicar['modelo']. map(lambda x: x.split()[0].upper()) #Nos quedamos solo con el modelo 
flexicar['kilometros']=flexicar['kilometros'].map(lambda x: int(float(x[:-3])*1000) if "." in ((x[:-3]))  else int((x[:-3])))
flexicar['precio']=flexicar['precio'].map(lambda x: int(float(x[:-1])*1000) if "." in ((x[:-1]))  else int((x[:-1])))
flexicar['marca']=flexicar['marca'].map(lambda x: x.upper())  #Escrbimos la marca en mayuscula 

flexicar
```

Limpieza csv ocasionplus

```python
ocasion_plus['modelo']=ocasion_plus['modelo'].map(lambda x: x.split()[0].upper())
ocasion_plus['kilometros']=ocasion_plus['kilometros'].map(lambda x: int(float(x[:-3])*1000) if "." in ((x[:-3]))  else int((x[:-3])))
ocasion_plus['precio']=ocasion_plus['precio'].map(lambda x: int(float(x[:-1])*1000) if "." in ((x[:-1]))  else int((x[:-1])))
ocasion_plus['marca']=ocasion_plus['marca'].map(lambda x: x.upper()) #Escrbimos la marca en mayuscula 
```

Limpieza csv Holanda


```python
Holanda['modelo']=Holanda['modelo'].map(lambda x: x.split()[0].upper()) #Nos quedamos solo con el modelo
Holanda['kilometros']=Holanda['kilometros'].map(lambda x : int((x.replace(",","")[:-2])))  #Limpieza de los kilometros
Holanda['precio']=Holanda['precio'].map(lambda x: int((x[:-1]).replace(",","") ) if "," in str(x[:-1])  else int(x[:-1])) #Limpieza del precio
Holanda['marca']=Holanda['marca'].map(lambda x: x.upper()) #Escrbimos la marca en mayuscula 
Holanda['marca'].replace('VOLSKWAGEN','VOLKSWAGEN', inplace=True) #Cambiamos un nombre mal escrito
Holanda['pais']='Holanda'
Holanda.to_csv('../data/processed/cars_netherlands.csv', index=False) #Pasamos a un nuevo csv el csv limpo
```

```python
Alemania['año']=Alemania['año'].map(lambda x : x.replace(",","")) #Quitamos  una coma de los años
Alemania.rename(columns={'km':'kilometros'},inplace=True) #Renombramos una columna 
Alemania['kilometros']=Alemania['kilometros'].map(lambda x : int(float(x[:-2])*1000) if "." in ((x[:-2]))  else int((x[:-2]))) #Limpieza de los kilometros
Alemania['precio']=Alemania['precio'].map(lambda x :int(float(x[:-9])*1000) if "." in ((x[:-9]))  else int((x[:-9])))  #Limpieza del precio
Alemania['marca']=Alemania['marca'].map(lambda x: x.upper()) #Escrbimos la marca en mayuscula 
Alemania['modelo']=Alemania['modelo'].map(lambda x: str(x).upper()) #Escribimos el modelo en mayusculas
Alemania['pais']='Alemania'

Alemania.to_csv('../data/processed/cars_germany.csv', index=False) #Pasamos a un nuevo csv el csv limpio
```

Limpieza de csv autocar

```python
autocasion.rename(columns={'a�o': 'año'}, inplace=True) #Renombramos una columna 
autocasion['marca']=autocasion['marca'].map(lambda x: x.upper()) #Escrbimos la marca en mayuscula 
autocasion['modelo']=autocasion['modelo'].map(lambda x: x.split()[0].upper()) #Nos quedamos solo con el modelo 
autocasion['precio']=autocasion['precio'].map(lambda x :int(float(x[:6])*1000) if "." in ((x[:6]))  else int((x[:6]))) #Limpiamos el precio
autocasion['kilometros']=autocasion['kilometros'].map(lambda x : int(float(x[:-2])*1000) if "." in ((x[:-2]))  else int((x[:-2]))) #Limpiamos los kilometros
```

Juntamos los csv de paginas españolas

```python
España=pd.concat([flexicar,cliclar,ocasion_plus,autocasion],ignore_index=True) #Juntamos csv de España
España['pais']='España'
España.to_csv('../data/processed/cars_spain.csv', index=False)#Escribmos en un nuevo csv el csv limpio
```
