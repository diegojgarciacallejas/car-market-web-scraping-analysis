```python
import pandas as pd
import matplotlib.pyplot as plt
#Leemos los csv
coches_francia_df= pd.read_csv('../data/processed/cars_france.csv', index_col= 0)
coches_francia_df['pais'] = 'Francia'
coches_holanda_df= pd.read_csv('../data/processed/cars_netherlands.csv', index_col= 0)
coches_holanda_df['pais'] = 'Holanda'
coches_alemania_df= pd.read_csv('../data/processed/cars_germany.csv', index_col= 0)
coches_alemania_df['pais'] = 'Alemania'
coches_españa_df= pd.read_csv('../data/processed/cars_spain.csv' ,index_col= 0)
coches_españa_df['pais'] = 'España'
```

```python
#Cogemos todos los renault 
df_renault_francia= coches_francia_df.loc['RENAULT'] 
df_renault_alemania= coches_alemania_df.loc['RENAULT']
df_renault_holanda= coches_holanda_df.loc['RENAULT']
df_renault_españa= coches_españa_df.loc['RENAULT']
#Unimos todos los renault en un mismo data frame

df_renault_todos= pd.concat([df_renault_francia, df_renault_alemania, df_renault_holanda, df_renault_españa])
```

```python
#Lo agrupamos por pais
unidades_renault_por_pais = df_renault_todos.groupby('pais').size()


unidades_renault_por_pais.plot(kind='barh')

plt.title('RENAULT CLIO ESPIRIT ALPINE')
plt.xlabel('Unidades')
plt.ylabel('País')
plt.minorticks_on()
plt.show()
```

```python
#Cogemos todos los seat
df_seat_francia= coches_francia_df.loc['SEAT']
df_seat_alemania= coches_alemania_df.loc['SEAT']
df_seat_holanda= coches_holanda_df.loc['SEAT']
df_seat_españa= coches_españa_df.loc['SEAT']
#Unimos todos los seat en un mismo data frame
df_seat_todos= pd.concat([df_seat_francia, df_seat_alemania, df_seat_holanda, df_seat_españa])
```

```python
#Lo agrupamos por pais

unidades_seat_por_pais = df_seat_todos.groupby('pais').size()

unidades_seat_por_pais.plot(kind='barh')

plt.title('SEAT IBIZA FR')
plt.xlabel('Unidades')
plt.ylabel('País')
plt.minorticks_on()
```

```python
#Cogemos todos los opel
df_opel_francia= coches_francia_df.loc['OPEL']
df_opel_alemania= coches_alemania_df.loc['OPEL']
df_opel_holanda= coches_holanda_df.loc['OPEL']
df_opel_españa= coches_españa_df.loc['OPEL']

#Unimos todos los opel en un mismo data frame
df_opel_todos= pd.concat([df_opel_francia, df_opel_alemania, df_opel_holanda,df_opel_españa])
```

```python
#Lo agrupamos por pais
unidades_opel_por_pais = df_opel_todos.groupby('pais').size()

unidades_opel_por_pais.plot(kind='barh')

plt.title('OPEL CORSA GS-LINE')
plt.xlabel('Unidades')
plt.ylabel('País')
plt.minorticks_on()
```

```python
#Cogemos todos los hyundai
df_hyundai_francia= coches_francia_df.loc['HYUNDAI']
df_hyundai_alemania= coches_alemania_df.loc['HYUNDAI']
df_hyundai_holanda= coches_holanda_df.loc['HYUNDAI']
df_hyundai_españa= coches_españa_df.loc['HYUNDAI']
#Unimos todos los hyundai en un mismo data frame
df_hyundai_todos= pd.concat([df_hyundai_francia, df_hyundai_alemania, df_hyundai_holanda, df_hyundai_españa])
```

```python
#Lo agrupamos por pais
unidades_hyundai_por_pais = df_hyundai_todos.groupby('pais').size()

unidades_hyundai_por_pais.plot(kind='barh')

plt.title('HYUNDAI I20 N-LINE')
plt.xlabel('Unidades')
plt.ylabel('País')
plt.minorticks_on()
```

```python
#Cogemos todos los toyata
df_toyota_francia= coches_francia_df.loc['TOYOTA']
df_toyota_alemania= coches_alemania_df.loc['TOYOTA']
df_toyota_holanda= coches_holanda_df.loc['TOYOTA']
df_toyota_españa= coches_españa_df.loc['TOYOTA']
#Unimos todos los toyota en un mismo data frame

df_toyota_todos= pd.concat([df_toyota_francia, df_toyota_alemania, df_toyota_holanda, df_toyota_españa])
```

```python
#Lo agrupamos por pais

unidades_toyota_por_pais = df_toyota_todos.groupby('pais').size()

unidades_toyota_por_pais.plot(kind='barh')

plt.title('TOYOTA YARIS GR-SPORT')
plt.xlabel('Unidades')
plt.ylabel('País')
plt.minorticks_on()
```

```python
#Cogemos todos los volkswagen

df_volkswagen_francia = coches_francia_df.loc['VOLKSWAGEN']
df_volkswagen_alemania = coches_alemania_df.loc['VOLKSWAGEN']
df_volkswagen_holanda = coches_holanda_df.loc['VOLKSWAGEN']
df_volkswagen_españa = coches_españa_df.loc['VOLKSWAGEN']
#Unimos todos los volkswagen en un mismo data frame

df_volkswagen_todos= pd.concat([df_volkswagen_francia, df_volkswagen_alemania, df_volkswagen_holanda, df_volkswagen_españa])
```

```python
#Lo agrupamos por pais
unidades_volks_por_pais = df_volkswagen_todos.groupby('pais').size()

unidades_volks_por_pais.plot(kind='barh')

plt.title('VOLKSWAGEN POLO R-LINE')
plt.xlabel('Unidades')
plt.ylabel('País')
plt.minorticks_on()
```

```python
#Unimos las graficas y cambiamos colores

colors = ['#0fd8aa', '#f84c27', '#8220f0', '#6df127']

plt.figure(figsize=(24, 10))


ax1 = plt.subplot(2, 3, 1)
unidades_renault_por_pais.plot(kind='barh', color=colors, ax=ax1)
ax1.set_title('RENAULT CLIO ESPIRIT ALPINE')
ax1.set_ylabel('') 
plt.minorticks_on()


ax2 = plt.subplot(2, 3, 2)
unidades_seat_por_pais.plot(kind='barh', color=colors, ax=ax2)
ax2.set_title('SEAT IBIZA FR')
ax2.set_ylabel('')
plt.minorticks_on()


ax3 = plt.subplot(2, 3, 3)
unidades_opel_por_pais.plot(kind='barh', color=colors, ax=ax3)
ax3.set_title('OPEL CORSA GS-LINE')
ax3.set_ylabel('')
plt.minorticks_on()


ax4 = plt.subplot(2, 3, 4)
unidades_hyundai_por_pais.plot(kind='barh', color=colors, ax=ax4)
ax4.set_title('HYUNDAI I20 N-LINE')
ax4.set_ylabel('')
plt.minorticks_on()


ax5 = plt.subplot(2, 3, 5)
unidades_toyota_por_pais.plot(kind='barh', color=colors, ax=ax5)
ax5.set_title('TOYOTA YARIS GR-SPORT')
ax5.set_ylabel('')
plt.minorticks_on()


ax6 = plt.subplot(2, 3, 6)
unidades_volks_por_pais.plot(kind='barh', color=colors, ax=ax6)
ax6.set_title('VOLKSWAGEN POLO R-LINE')
ax6.set_ylabel('')
plt.minorticks_on()


plt.suptitle('Número de unidades por país y modelo de coche', fontsize=20, fontweight='bold')

plt.show()
```

```python
#Unimos todos los sublot en una misma grafica 
df_todos = pd.concat([df_opel_todos,df_seat_todos,df_toyota_todos,df_hyundai_todos,df_renault_todos,df_volkswagen_todos])
counts = df_todos.groupby(['marca', 'pais']).size().unstack(fill_value=0)
colors = ['#0fd8aa', '#f84c27', '#8220f0', '#6df127']
counts.plot(kind='bar', figsize=(10, 5), width=0.5, color = colors)
plt.title('Número de unidades por marca y país', fontsize=14)
plt.xlabel('Marca', fontsize=12)
plt.ylabel('Número de unidades', fontsize=12)
plt.xticks(rotation=45, fontsize=10)
plt.yticks(fontsize=10)
plt.legend(title='País', fontsize=10)
plt.grid(axis='y', linestyle='--', alpha=0.7)
```
